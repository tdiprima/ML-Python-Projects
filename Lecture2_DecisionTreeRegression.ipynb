```python
import pandas as pd
import numpy as np
import json
```

```python
from sklearn import datasets ## imports datasets from scikit-learn
from sklearn.model_selection import train_test_split
from sklearn import model_selection
from sklearn.tree import DecisionTreeRegressor  #decision tree regression
from sklearn.metrics import accuracy_score
```

```python
data = datasets.load_boston() ## loads Boston dataset from datasets library 
# parsed = json.loads(data)
# print(json.dumps(parsed, indent=4, sort_keys=True))
# def pp_json(json_thing, sort=True, indents=4):
#    if type(json_thing) is str:
#        print(json.dumps(json.loads(json_thing), sort_keys=sort, indent=indents))
#    else:
#        print(json.dumps(json_thing, sort_keys=sort, indent=indents))
#    return None
# pp_json(data)
data
```

```python
# define the data/predictors as the pre-set feature names  
X = pd.DataFrame(data.data, columns=data.feature_names)
X
```

```python
# Put the target (housing value -- MEDV) in another DataFrame
Y = pd.DataFrame(data.target, columns=["MEDV"])
# data.target
Y
```

```python
# SPLIT THE DATA
# sklearn.model_selection.train_test_split(*arrays, **options)[source]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = .25, random_state=25) #25% hold out for testing
print('X_train')
print(X_train)
print('X_test')
print(X_test)
print('y_train')
print(y_train)
print('y_test')
print(y_test)
```

```python
# instantiate decision tree model (depth = 2)
# if the maximum depth of the tree (controlled by the max_depth parameter) is set too high, overfitting may happen

regressor = DecisionTreeRegressor(max_depth=2)
```

```python
# FIT = Build a decision tree regressor from the training set (X, y).
regressor.fit(X_train, y_train)
```

```python
# predict Y values- 25% training data
Y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, Y_pred) #MSE
```
